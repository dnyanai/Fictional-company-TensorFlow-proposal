```
import random
import numpy as np
import os
import glob
import cv2
#from features import *
```

```
def resize(path, img_height, img_width):
    img = cv2.imread(path)
    resized = cv2.resize(img, (img_height, img_width))
    return resized
```

```
def get_data():
    train_dir = "/content/state-farm-dataset/imgs/train"
    X_train = []
    y_train = []
    
    for j in range(10):
        print('Load folder c{}'.format(j))
        path = os.path.join(train_dir, 'c' + str(j), '*.jpg')
        files = glob.glob(path)
        for fl in files:
            flbase = os.path.basename(fl)
            img = resize(fl, 64, 64)
            X_train.append(img)
            y_train.append(j)
    return X_train, y_train
```

```
X_train, y = get_data()
```

```
X_train = np.asarray(X_train)
y = np.asarray(y)
print(X_train.shape)
```

```
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=0.2, random_state=42)
```

```
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_val = np.reshape(X_val, (X_val.shape[0], -1))
print('Training data shape: ', X_train.shape)
print('Validation data shape: ', X_val.shape)
```

```
from sklearn.neural_network import MLPClassifier
lr = 1e-3
clf = MLPClassifier( solver='adam', alpha=5e-6, batch_size=200, max_iter=20,learning_rate_init=lr,
                     hidden_layer_sizes=(500,4), random_state=1,verbose=True)

clf.fit(X_train, y_train)
#MLPClassifier(alpha=1e-05, hidden_layer_sizes=(5, 2), random_state=1,
#              solver='lbfgs')
```

```
clf.score(X_val,y_val)
```
